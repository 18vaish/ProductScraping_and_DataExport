```python
import requests
from bs4 import BeautifulSoup

# Define the URL template with placeholders for page number
base_url = "https://www.amazon.in/s?k=bags&crid=2M096C61O4MLT&qid=1653308124&sprefix=ba%2Caps%2C283&ref=sr_pg_1"
# Number of pages to scrape
num_pages = 20

# Initialize empty lists to store scraped data
product_urls = []
product_names = []
product_prices = []
ratings = []
review_counts = []

# Loop through each page
for page_num in range(1, num_pages + 1):
    url = base_url.format(page_num)
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    # Extract product information
    products = soup.find_all("div", {"data-component-type": "s-search-result"})

    for product in products:
        try:
            product_url = "https://www.amazon.in" + product.find("a", class_="a-link-normal")["href"]
            product_name = product.find("span", class_="a-size-medium").text.strip()
            product_price = product.find("span", class_="a-price-whole").text.strip()
            rating = product.find("span", class_="a-icon-alt").text.split()[0]
            review_count = product.find("span", class_="a-size-base").text.strip()

            # Append data to respective lists
            product_urls.append(product_url)
            product_names.append(product_name)
            product_prices.append(product_price)
            ratings.append(rating)
            review_counts.append(review_count)
        except:
            pass

# Print or save the scraped data
for i in range(len(product_urls)):
    print("Product URL:", product_urls[i])
    print("Product Name:", product_names[i])
    print("Product Price:", product_prices[i])
    print("Rating:", ratings[i])
    print("Number of Reviews:", review_counts[i])
    print("-" * 50)

# You can also save the data to a CSV or other formats if needed
```

```python
import requests
import csv
from bs4 import BeautifulSoup
from time import sleep

# ... (Rest of the code remains the same until the loop)

# Loop through each product URL to fetch additional information
for product_url in product_urls:
    response = requests.get(product_url)
    soup = BeautifulSoup(response.content, 'html.parser')

    try:
        description = soup.find("span", {"id": "productTitle"}).get_text(strip=True)
    except AttributeError:
        description = "N/A"
    descriptions.append(description)

    try:
        asin = soup.find("th", string="ASIN").find_next("td").get_text(strip=True)
    except AttributeError:
        asin = "N/A"
    asins.append(asin)

    try:
        product_description = soup.find("div", {"id": "productDescription"}).get_text(strip=True)
    except AttributeError:
        product_description = "N/A"
    product_descriptions.append(product_description)

    try:
        manufacturer = soup.find("a", {"id": "bylineInfo"}).get_text(strip=True)
    except AttributeError:
        manufacturer = "N/A"
    manufacturers.append(manufacturer)

    sleep(2)  # Add a delay between requests

# Save the data to a CSV file
csv_filename = "amazon_products.csv"

with open(csv_filename, mode="w", newline="", encoding="utf-8") as csv_file:
    csv_writer = csv.writer(csv_file)
    csv_writer.writerow(["Product URL", "Product Name", "Product Price", "Rating", "Number of Reviews", "Description", "ASIN", "Product Description", "Manufacturer"])

    for i in range(len(product_urls)):
        csv_writer.writerow([product_urls[i], product_names[i], product_prices[i], ratings[i], review_counts[i],
                            descriptions[i], asins[i], product_descriptions[i], manufacturers[i]])

print("Data has been exported to", csv_filename)
```
